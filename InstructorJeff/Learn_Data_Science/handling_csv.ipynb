```python
#Use built-in csv module
import csv

#open csv file
with open('data.csv', 'r') as csvfile:
    #create a csv reader object
    csvreader = csv.reader(csvfile)
    
    #read the header row
    header = next(csvreader)
    print("Header:", header)
    
    #read the data rows
    for row in csvreader:
        print("Row:", row)
        
    
```
