```python
#Data Visualization with Matplotlib - quickest way to make sense of your data
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


df = pd.read_csv('data.csv')

#Histogram of Duration
"""
1. set up the figure size
2. create a histogram of the 'Duration' column
"""
plt.figure(figsize=(10, 6))
plt.subplot(2,2,1)
sns.histplot(df['Duration'], bins=30, kde=True, color='blue')
plt.title('Histogram of Duration')
```

```python
#scatter plot of Duration vs. Calories
plt.figure(figsize=(15, 10))  #15 and 5 are the height and width of the figure
subplot = plt.subplot(2,2,1)
sns.scatterplot(x='Duration', y='Calories', data=df, color='red')
plt.title('Scatter Plot of Duration vs. Calories')

#scatter plot of pulse
plt.subplot(2,2,2)
sns.scatterplot(x='Pulse', y='Calories', data=df, color='green')
plt.title('Scatter Plot of Pulse vs. Calories')

```

```python
#correlation heatmap
corr = df.corr() #correlation matrix
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', square=True)
plt.title('Correlation Heatmap')
plt.show()
```
