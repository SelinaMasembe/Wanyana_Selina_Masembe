```python
#read csv file
import pandas as pd
df =pd.read_csv('AfricaCupofNationsMatches.csv')
```

```python
#getting first 7rows
print(df.head(7))
```

```python
#select home team,away team,home team goals,away team goals
df1 = df[['HomeTeam', 'AwayTeam', 'HomeTeamGoals', 'AwayTeamGoals']]
#displaying first 7 rows of selected columns
print(df1.head(7))
```

```python
#select rows where Egypt appears as home team
df2 = df1[df1['HomeTeam'] == 'Egypt']
print("Egypt as Home Team:")
print(df2)
print('\n')

df3 = df1[df1['AwayTeam'] == 'Egypt']
print("Egypt as Away Team:")        
print(df3)


```

```python
#rows with Egypt
egypt_rows = df1[(df1['HomeTeam'] == 'Egypt') | (df1['AwayTeam'] == 'Egypt')]
# Displaying rows with Egypt as either home or away team
print("Rows with Egypt as either Home or Away Team:")
print(egypt_rows)


```

```python
#count rows and columns of dataframe
print("Number of rows and columns in the dataframe:", df1.shape)
```

```python
#select all rows where attendance is missing
df4 = df[df['Attendance'].isnull()]
df5 = df4[['HomeTeam', 'AwayTeam', 'Stage','Stadium','Attendance']]
print("Rows with missing attendance:")
print(df5)
```

```python
#select roes where home team goals are between 3 and 6 inclusive
df6 = df1[(df1['HomeTeamGoals'] >= 3) & (df1['HomeTeamGoals'] <= 6)]
print("Home team goals between 3 and 6 inclusive:")
print(df6)
```

```python
#change the “AwayTeamGoals” in the 3rd row to 10
print ("Original AwayTeamGoals in the 3rd row:")
print(df.at[2, 'AwayTeamGoals'])
```

```python
df.at[2, 'AwayTeamGoals'] = 10 #at[2, 'AwayTeamGoals'] accesses the 'AwayTeamGoals' column in the 3rd row
print("Updated AwayTeamGoals in the 3rd row:")
print(df.iloc[2])
```

```python
#sort the DataFrame first by ‘HomeTeam’ in ascending order
df_sorted = df1.sort_values(by='HomeTeam', ascending=True)
print("DataFrame sorted by HomeTeam in ascending order:")
print(df_sorted.head(25))
```

```python
#sort dataframe by HomeTeamScores’ in descending order
df_sorted_desc = df1.sort_values(by='HomeTeamGoals', ascending=False)
print("DataFrame sorted by HomeTeamGoals in descending order:") 
print(df_sorted_desc.head(25))
```

```python
#get list from DataFrame column headers.
column_headers = df.columns.tolist()
print("List of DataFrame column headers:")
print(column_headers)

```

```python
# append a column at a specific position, e.g., after the 'HomeTeam' column.
venues = (['Venue1', 'Venue2', 'Venue3', 'Venue4', 'Venue5', 'Venue6', 'Venue7'] * ((len(df) // 7) + 1))[:len(df)]
df.insert(1, 'Venue', venues)  # Insert a new column at position 1 with correct length
print("DataFrame with 'Venue' column inserted after 'HomeTeam':")
print(df.head(7))
```

```python
#changing order of data frame columns
print(df.columns)
new_column_order = ['Venue', 'HomeTeam', 'AwayTeam', 'HomeTeamGoals', 'AwayTeamGoals', 'Referee', 'Stage', 'Stadium', 'Attendance']
df = df[new_column_order]
print('\n')
print("DataFrame with columns reordered:")
print(df.columns)
```

```python
# To add 2 rows to your DataFrame with all columns filled with values of your choice.
# First, add the missing columns to df if they do not exist
for col in ['Year', 'Date ', 'City']:
    if col not in df.columns:
        df[col] = pd.NA

# Prepare new rows with all columns present in df
new_rows = pd.DataFrame({
    'Venue': ['VenueX', 'VenueY'],
    'HomeTeam': ['TeamA', 'TeamB'],
    'AwayTeam': ['TeamC', 'TeamD'],
    'HomeTeamGoals': [2, 3],
    'AwayTeamGoals': [1, 0],
    'Referee': ['RefereeX', 'RefereeY'],
    'Stage': ['Group Stage', 'Quarter Finals'],
    'Stadium': ['Namboole', 'Hoima'],
    'Attendance': [50000, 60000],
    'Year': [2023, 2024],
    'Date ': [pd.to_datetime('2023-01-01'), pd.to_datetime('2024-01-01')],
    'City': ['Kampala', 'Hoima']
})

# Reorder columns to match df
new_rows = new_rows[df.columns]

# Append the new rows to the DataFrame
df = pd.concat([df, new_rows], ignore_index=True)
print("DataFrame after adding new rows:")
print(df.tail(2))
# Reorder columns to match df
new_rows = new_rows[df.columns]

# Append the new rows to the DataFrame
df = pd.concat([df, new_rows], ignore_index=True)
```
